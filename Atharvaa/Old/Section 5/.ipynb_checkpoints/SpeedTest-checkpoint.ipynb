```python
#Using Python Lists
import numpy as np
import time

size = 1_000_000

L1 = list(range(size))
L2 = list(range(size))

start = time.time()
add = [x+y for x,y in zip(L1, L2)]
end = time.time()
print(add[0:100])

print(end - start)
```

```python
#Using Numpy Arrays
import numpy as np
import time

size = 1_000_000_0

L1 = np.array(range(size))
L2 = np.array(range(size))

start = time.time()
add = L1 + L2
end = time.time()
print(add[0:100])

print(end - start)
```
